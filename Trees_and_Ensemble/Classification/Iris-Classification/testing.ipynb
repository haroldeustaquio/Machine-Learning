# 1. Importing Libraries


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix, f1_score, precision_score,recall_score
from sklearn.datasets import load_iris
```

# 2. Loading and Exploring the Data

```python
x = pd.DataFrame(load_iris()['data'], columns = load_iris()['feature_names'])
y = load_iris()['target']
```

```python
x_train,x_test , y_train, y_test = train_test_split(x,y, test_size=0.3)
```

# 5. Model Building and Training


```python
model_1 = DecisionTreeClassifier(criterion='gini')

model_1.fit(x_train, y_train)
predictions_1 = model_1.predict(x_test)
```

```python
plot_tree(model_1)
```

```python
model_2 = DecisionTreeClassifier(criterion='entropy',max_depth=5)

model_2.fit(x_train, y_train)
predictions_2 = model_2.predict(x_test)
```

```python
plot_tree(model_2)
```

# 6. Model Evaluation

```python
confusion_matrix(y_test,predictions_1)
```

```python
confusion_matrix(y_test, predictions_2)
```
